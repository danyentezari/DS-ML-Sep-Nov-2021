### Import Data and Libraries

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import GridSearchCV
```

```python
housesDf = pd.read_csv('./ml_house_data_set.csv')
housesDf.head()
```

```python
irrelevantFeatures = ['house_number', 'street_name', 'unit_number', 'zip_code']
housesDf.drop( irrelevantFeatures, axis=1, inplace=True )
housesDf.head()
```

```python
# Apply one-hot encoding to the categorical data
housesDf = pd.get_dummies( housesDf, columns=['garage_type', 'city'] )
```

```python
housesDf.info()
```

```python
# Select the features and label
labelDf = housesDf['sale_price']
featuresDf = housesDf.drop('sale_price', axis=1, inplace=False)
```

```python
# Train, Validation, Test Split
trainF, testF, trainL, testL = train_test_split( featuresDf, labelDf, test_size=0.4, random_state=33 )
```

```python
gbModel = GradientBoostingRegressor(
    n_estimators=1000,          # Number of decision trees
    learning_rate=0.1,          # How much time is spent learning the parameters
    max_depth=6,                # Max depth for any tree in the forest
    min_samples_leaf=9,         # Minimum number of types of observations a tree should learn from 
    max_features=0.1,           # Maximum percentage of features any tree should learn from 
    loss='huber'                # The function used for calculating the errors of the model
)
```

```python
gbModel.fit( trainF, trainL )
```

```python
mae_train = mean_absolute_error( trainL, gbModel.predict(trainF) )
print( mae_train )
```

```python
mae_test = mean_absolute_error( testL, gbModel.predict(testF) )
print( mae_test )
```
