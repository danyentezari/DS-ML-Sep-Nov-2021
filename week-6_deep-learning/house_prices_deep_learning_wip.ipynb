# Deep Learning for House Prices

### Import Data and Libraries

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import *
from sklearn.model_selection import train_test_split
```

```python
houseDf = pd.read_csv('./ml_house_data_set_scaled.csv')
houseDf.head()
```

```python
featuresDf = houseDf.drop( 'sale_price', axis=1, inplace=False )
labelDf = houseDf['sale_price']
```

### Train-Test Split

```python
trainF, testF, trainL, testL =  train_test_split( featuresDf, labelDf, test_size=0.4, random_state=1 )
```

### Build the Neural Network for Deep Learning

```python
modelDL = Sequential()

# Input Layer
modelDL.add( Dense(50, input_dim=63, activation='relu') )

# Hidden Layers
modelDL.add( Dense(100, activation='relu') )
modelDL.add( Dense(50, activation='relu') )

# Output Layer
modelDL.add( Dense(1, activation='linear') )

# Compile the neural network
modelDL.compile( loss='mean_squared_error', optimizer='adam' )
```
